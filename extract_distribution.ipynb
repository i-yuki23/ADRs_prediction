```python
from datasets import load_dataset, DatasetDict
import torch
import pandas as pd 
import os

data_dir = "./data/processed_data/scaffold/"
# Load the datasets

train_dataset = pd.read_csv(os.path.join(data_dir, "train/metadata.csv"))
```

```python
train_dataset
```

```python
train_dataset.drop(0)
```

```python
train_dataset.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
class_names = train_dataset.columns
print(class_names)
print("num_labels:", len(class_names))
print(class_names[0])
```

```python
class_distributions = []

for class_name in class_names[1:]:
    class_distribution = train_dataset[class_name].value_counts()
    class_distributions.append(class_distribution)
print(class_distributions)
```

```python
class_dist_rate = []
for class_dist in class_distributions:
    class_rate = class_dist[0]/class_dist[1]
    class_dist_rate.append(class_rate)

print(class_dist_rate)

# Convert the list to a tensor
tensor = torch.tensor(class_dist_rate)

# Save the tensor to a file
torch.save(tensor, './data/processed_data/train_class_dist_rate.pt')
```
